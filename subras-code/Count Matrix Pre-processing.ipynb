```python
import pandas as pd
import numpy as np
```

```python
# Segerstolpe pancreas data
RNA_DATA_PATH = 'segerstolpe_pancreas_counts_3514sc.csv'
CELLTYPE_DATA_PATH = 'segerstolpe_pancreas_annotations.csv'

df_GE = pd.read_csv(RNA_DATA_PATH, index_col = 0)
df_LABEL = pd.read_csv(CELLTYPE_DATA_PATH, index_col = 0)
```

```python
df_GE = df_GE.reindex(sorted(df_GE.columns), axis = 1)
```

```python
df_GE.shape
```

```python
df_GE = df_GE[~df_GE.index.duplicated(keep = 'first')]
df_GE['counts'] = df_GE.sum(axis = 1)
df_GE = df_GE[df_GE.counts > 10]
#df_GE = df_GE.iloc[:,:-1].div(df_GE.counts, axis = 0)
```

```python
df_GE.head()
```

```python
df_GE.loc['Total'] = df_GE.sum(axis = 0)
```

```python
df_GE = df_GE.iloc[:-1,:].div(df_GE.iloc[-1][:-1])*100000
```

```python
df_LABEL = df_LABEL.rename(columns = {'Characteristics[cell type]':'truth'})
badtypes = df_LABEL[df_LABEL.truth.isin(['not applicable','unclassified cell'])].index
df_LABEL = df_LABEL[~df_LABEL.index.isin(badtypes)]
```

```python
df_GE = df_GE.loc[:,~df_GE.columns.isin(badtypes)]
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
le.fit(df_LABEL.truth)
mapping = pd.DataFrame({'Cell Type':le.classes_, 'Encoding':le.transform(le.classes_)})
df_LABEL.truth = le.fit_transform(df_LABEL.truth)
```

```python
df_GE.iloc[:,:-1].to_csv('segerstolpe_pancreas_counts_clean2.csv')
df_LABEL.to_csv('segerstolpe_pancreas_annotations_clean2.csv')
mapping.to_csv('segerstolpe_annotation_map2.csv')
```
